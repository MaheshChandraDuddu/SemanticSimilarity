```python
flip_data['info'] = flip_data['product_name'] + " " + flip_data['description']
amz_data['info'] = amz_data['product_name'] + " " + amz_data['description']
```

```python
!pip install transformers
!pip install sentence-transformers
from sentence_transformers import SentenceTransformer, util
```

```python
model = SentenceTransformer('stsb-roberta-large')
```

```python
st1 = flip_data['info'].values.tolist()
st2 = amz_data['info'].values.tolist()
embedding1 = model.encode(st1, convert_to_tensor=True)
embedding2 = model.encode(st2, convert_to_tensor=True)
```

```python
from tqdm import tqdm
```

```python
#This takes a lot of time to run and hence the outer for loop range can be decreased
df = pd.DataFrame()
for i in tqdm(range(len(flip_data))):
  max_ind = 0
  max_score = -1
  st1 = str(flip_data.loc[i, 'info'])
  embed1 = embedding1[i]
  for j in (range(len(amz_data))):
    st2 = str(amz_data.loc[j, 'info'])
    embed2 = embedding2[j]
    cosine_scores = util.pytorch_cos_sim(embed1, embed2)
    score = cosine_scores.item()
    if score >= max_score:
      max_score = score
      max_ind = j
  df.loc[i, 'Product name in Flipkart'] = flip_data.loc[i, 'product_name']
  df.loc[i, 'Flipkart Retail Price'] = flip_data.loc[i, 'retail_price']
  df.loc[i, 'Flipkart Discounted Price'] = flip_data.loc[i, 'discounted_price']
  df.loc[i, 'Product name in Amazon'] = amz_data.loc[max_ind, 'product_name']
  df.loc[i, 'Amazon Retail Price'] = amz_data.loc[max_ind, 'retail_price']
  df.loc[i, 'Amazon Discounted Price'] = amz_data.loc[max_ind, 'discounted_price']
```
